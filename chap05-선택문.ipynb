[예제 5-1] 입력받은 양수가 홀수 또는 짝수인지 확인하기

```C
#include <stdio.h>

int main(void)
{
    int input;
    printf("양수를 입력하세요.\n");
    scanf("%d", &input);

    if (input%2 == 0)
        printf("입력한 수 %d는 짝수입니다.\n", input);
}
```

````python
%%write to ex5_1.c
#include <stdio.h>

int main(void)
{
    int input;
    printf("양수를 입력하세요.\n");
    scanf("%d", &input);

    if (input%2 == 0)
        printf("입력한 수 %d는 짝수입니다.\n", input);
}
````
