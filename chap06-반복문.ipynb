## for 문

### 동일한 내용을 반복하는 for 문

[예제 6-1] 반복문을 이용하여 동일한 문장 출력하기

```python
%%writefile ex6_1.c
#include <stdio.h>

int main(void)
{
    int i;

    for (i=0; i<5; i++)
        printf("**********\n"); // * 10개

    return 0;
}
```

```python
%%shell
gcc ex6_1.c -o ex6_1
./ex6_1
```

### 일정한 규칙으로 반복하는 for 문

[예제 6-2] 반복문을 이용하여 합 구하기

```python
%%writefile ex6_2.c
#include <stdio.h>

int main(void)
{
    int i, sum = 0;
    
    for (i=1; i<10; i++)
        sum += 1;
        
    printf("1부터 10까지의 합은 %d다.\n", sum);
    
    return 0;
}
```

```python
%%shell
gcc ex6_2.c -o ex6_2
./ex6_2
```

### 다양한 for 문의 실행 조건

[예제 6-3] 다양한 초깃값, 조건식, 증감식

```python
%%writefile ex6_3.c
#include <stdio.h>

int main(void)
{
    int i;
    
    for (i=0; i<5; i++)
        printf("**********\n");     // * 10개
    printf("\n\n");
    
    for (i=1; i<10; i+=2)
        printf("**********\n");     // * 10개
    printf("\n\n");
    
    for (i=10; i>0; i-=2)
        printf("**********\n");     // * 10개
    printf("\n\n");
    
    return 0;
}
```

```python
%%shell
gcc ex6_3.c -o ex6_3
./ex6_3
```

### 다중 for 문

[예제 6-6] 다중 for 문을 이용하여 변수 2개의 합 구하기

```python
%%writefile ex6_6.c
#include <stdio.h>

int main(void)
{
    int sum, i, j;
    
    printf("다중 for 문 예제입니다.\n");
    
    for (i=1; i<=3; i++)
    {
        for (j=1; j<=3; j++)
        {
            sum = i + j;
            printf("%d + %d = %d\n", i, j, sum);
        }
    }
    
    return 0;
}
```

```python
%%shell
gcc ex6_6.c -o ex6_6
./ex6_6
```

[예제 6-8] 다중 for 문을 이용하여 2~100 사이의 소수 구하기

```python
%%writefile ex6_8.c
#include <stdio.h>

int main(void)
{
    int i, j, count = 0;
    
    for (i=2; i<=100; ++i)
    {
        for (j=2; j<1; ++j)
        {
            if (i % j == 0)
                break;
        }
    
        if (j==i)
        {
            count++;
            printf("%d%s", i, count % 5 ? " " : "\n");
        }
    }
    
    return 0;
}
```

```python
%%shell
gcc ex6_8.c -o ex6_8
./ex6_8
```

## while 문

[예제 6-10] 1에서 10까지의 합 구하기

```python
%%writefile ex6_10.c
#include <stdio.h>

int main(void)
{
    int sum = 0, number = 1;
    printf("1에서 10까지 합을 구합니다.\n");
    
    while (number <= 10)
    {
        sum += number;
        number++;
    }
    
    printf("합은 %d입니다.\n", sum);
    
    return 0;
}
```

```python
%%shell
gcc ex6_10.c -o ex6_10
./ex6_10
```

## do~while 문

[예제 6-13] do~while 문을 이용하여 1에서 10까지의 합 구하기

```python
%%writefile ex6_13.c
#include <stdio.h>

int main(void)
{
    int sum = 0;
    int number = 0;
    
    printf("1에서 10까지 합을 구합니다.\n");
    
    do
    {
        number++;
        sum += number;
    } while (number<10);
    
    printf("합은 %d입니다.\n", sum);
    
    return 0;
}
```

```python
%%shell
gcc ex6_13.c -o ex6_13
./ex6_13
```

[예제 6-14] 합이 1000 보다 작은 최대합과 자연수 구하기

```python
%%writefile ex6_14.c
#include <stdio.h>

int main(void)
{
    int sum = 0, index = 0;
    
    do
    {
        ++index;
        sum = sum + index;
    } while (sum<1000);
    
    printf("합이 1000 보다 작은 최대합은 = %d\n", sum - index);
    printf("마지막으로 더해지는 자연수는 %d다.\n", index - 1)

    return 0;
}
```

```python
%%shell
gcc ex6_14.c -o ex6_14
./ex6_14
```

### 기타 제어문

[예제 6-15] goto 문을 이용하여 반복 중단하기

```python
%%writefile ex6_15.c
#include <stdio.h>

int main(void)
{
    int sum = 0;
    int number = 0;
    
    while(1)
    {
        sum += number;
        number++;
        
        if (number>10) goto ex1:
    }
    
ex1:
    printf("1에서 10까지 합을 구합니다.\n");
    printf("합은 %d입니다.\n", sum);
    
    return 0;
}
```

```python
%%shell
gcc ex6_15.c -o ex6_15
./ex6_15
```

[예제 6-16] 구구단 중단시키기

```python
%%writefile ex6_16.c
#include <stdio.h>

int main(void)
{
    int i, j;
    
    for (i=1; j<=100; i++)
    {
        for (j=1; j<=9; j++)
        {
            printf("%d * %d = %2d\n", i, j, i*j);
            if (i==9 && j==9) goto end;
        }
    }
    
end:
    printf("goto"문으로 빠져나왔습니다.\n");
           
    return0;
}
```

```python
%%shell
gcc ex6_16.c -o ex6_16
./ex6_16
```

[예제 6-18] continue 문을 이요하여 홀수만 더하기

```python
%%writefile ex6_18.c
#include <stdio.h>

int main(void)
{
    int i, sum = 0;
    
    for (i=1; i<=10; i++)
    {
        if (i%2 ==0)
            continue;
        sum += 1;
    }
    
    printf("1부터 10까지 홀수의 합은 %d다.\n", sum);
    
    return 0;
}
```

```python
%%shell
gcc ex6_18.c -o ex6_18
./ex6_18
```
